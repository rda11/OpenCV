```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("Computer-Vision-with-Python/DATA/dog_backpack.png")
```

```python
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
flip = img_rgb.copy()
```

```python
flip = cv2.flip(flip,0)
plt.imshow(flip)
```

```python
cv2.rectangle(img_rgb,(200,380),(600,700),(255,0,0),7)
plt.imshow(img_rgb)
```

```python
vertices = np.array([[250,700],[480,400],[600,700]],np.int32)
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
cv2.polylines(img_rgb,[pts],isClosed=True,color=(0.0,255),thickness=20)
plt.imshow(img_rgb)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.fillPoly(img_rgb,[pts],color=(0.0,255))
plt.imshow(img_rgb)
```
