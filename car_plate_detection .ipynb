```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Computer-Vision-with-Python/DATA/car_plate.jpg')
```

```python
def display(img):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

```python
display(img)
```

```python
plate_cascade= cv2.CascadeClassifier('Computer-Vision-with-Python/DATA/haarcascades/haarcascade_russian_plate_number.xml')
```

```python
def detect_plate(img):
    plate_img= img.copy()
    plate_rects= plate_cascade.detectMultiScale(plate_img)
    
    for (x,y,w,h) in plate_rects:
        cv2.rectangle(plate_img,(x,y),(x+w,y+h),(0,255,0),4)
        
        return plate_img

    
    
```

```python
result = detect_plate(img)
```

```python
display(result)
```
